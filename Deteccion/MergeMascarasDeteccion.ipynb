```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2
import re
```

```python
csv = pd.read_csv(os.path.join('/home','wvillegas','dataset-mask','detection_dataset.csv'))
path_images = os.path.join('/home','wvillegas','dataset-mask','images')
```

```python
path_dest = os.path.join('/home','wvillegas','dataset-mask','masks')
```

```python
csv.columns
```

```python
orig = csv['imageOrigin'].unique()
```

```python
test = csv.loc[csv['imageOrigin'] == orig[217]]['mask'].iloc[3]
```

```python
csv.loc[csv['imageOrigin'] == orig[217]]['mask'].iloc[1]
```

```python
img_test = cv2.imread(os.path.join(path_images,test))
img2_test = cv2.imread(os.path.join(path_images,'0218_7983_mask.png'))
img3_test = cv2.imread(os.path.join(path_images,'0218_7982_mask.png'))
```

```python
cv2.imwrite("test.png",cv2.add(cv2.add(img_test,img3_test),img2_test))
```

```python
list_orig = []
list_maskfull = []
for img in orig:
    img_number = (re.findall(r'[0-9]+',img))[0]
    #print(img_number)
    df_temp = csv.loc[csv['imageOrigin'] == img]
    mask = cv2.imread(os.path.join(path_images, df_temp['mask'].iloc[0]))
    print(mask.shape)
    if(df_temp.shape[0] > 1):
        for idx in np.arange(df_temp.shape[0]):
            if (idx != 0):
                addition = cv2.imread(os.path.join(path_images, df_temp['mask'].iloc[idx]))
                mask = cv2.add(mask,addition)
    cv2.imwrite(os.path.join(path_dest,'mask_'+img_number+'.png'),mask)
    list_orig.append(img)
    list_maskfull.append('mask_'+img_number+'.png')
```

```python
dataframe = pd.DataFrame({
    'orig':list_orig,
    'mask':list_maskfull
})
dataframe.to_csv(os.path.join('/home','wvillegas','dataset-mask','full_masks.csv'))
```
