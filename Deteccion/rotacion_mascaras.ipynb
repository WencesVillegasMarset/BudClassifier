```python
new_path_img = path.join('/home','wvillegas','dataset-mask','dataset_resize','images_resize')
new_path_mask = path.join('/home','wvillegas','dataset-mask','dataset_resize','masks_resize')
```

```python
rows = []
for idx, row in dataset.iterrows():
    img = cv2.imread(path_images + '/' + row['orig'])
    msk = cv2.imread(path_mask + '/' + row['mask'])
    img = cv2.resize(img,(2000,2000))
    msk = cv2.resize(msk,(2000,2000))
    cv2.imwrite(new_path_img + '/' + row['orig'], img)
    cv2.imwrite(new_path_mask + '/' + row['mask'], msk)
```
